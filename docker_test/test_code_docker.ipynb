```python
import numpy as np
import matplotlib.pyplot as plt

# some sample data
x = np.linspace(0, 2 * np.pi, 100) 
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
